Mosphology has
- no apical dendrite:
> As mentioned in the exercice description, some of the interneurons have no dendrite destinction and are registered as not having apical dendrite on NeuroM. Therefore, it doesn't necessary mean there is an error
- jumps: 
> Jumps might be the results of precision of focus during microscopy imaging, which can be corrected manually based on morphological type of the cell and surrounding slices.
- fat ends:
>  A fat end is experiences when the end of the segment is larger than the mean of the rest by a factor of multiple_of_mean. This can be corrected by using other ends of the neurons within the same morphological class.
- narrow start:
> Indicates a narrower initial section of a neuron compared to the following section. This error can be corrected by examining the corresponding segments of the neurons with same morphological type.
- dangling branch:
> When a branch gets disconnected from the neuron, this can be solved automatically by connecting the branch to the nearest point on the neuron

Let's now compute bar plots of a variety of different characteristics, in order to understand which ones could be most relevant to our classification.

```python
from math import floor, ceil, sqrt
from neurom import NeuriteType
from neurom import features
characteristics = {
    "Axon volume densities": [features.morphology.neurite_volume_density(m, neurite_type=NeuriteType.axon)[0] for m in morphs],
    "Axon volumes": [features.morphology.total_volume_per_neurite(m, neurite_type=NeuriteType.axon)[0] for m in morphs],
    "Basal dendrite volume densities (avg)": [np.array(features.morphology.neurite_volume_density(m, neurite_type=NeuriteType.basal_dendrite)).mean() for m in morphs],
    "Basal dendrite volume (avg)": [np.array(features.morphology.total_volume_per_neurite(m, neurite_type=NeuriteType.basal_dendrite)).mean() for m in morphs],
    "Mean trunk section lenght": [np.array(features.morphology.trunk_section_lengths(m)).mean() for m in morphs],
    "Basal dendrite volume densities (max)": [np.array(features.morphology.neurite_volume_density(m, neurite_type=NeuriteType.basal_dendrite)).max() for m in morphs],
    # "Basal dendrite volumes (max)": [np.array(features.morphology.total_volume_per_neurite(m, neurite_type=NeuriteType.basal_dendrite)).max() for m in morphs],
    # "Apical dendrite volume densities": [features.morphology.neurite_volume_density(m, neurite_type=NeuriteType.apical_dendrite)[0] for m in morphs],
    # "Apical dendrite volumes": [features.morphology.total_volume_per_neurite(m, neurite_type=NeuriteType.apical_dendrite)[0] for m in morphs],
    "Total lenght per neurite": [np.array(features.morphology.total_length_per_neurite(m, neurite_type=NeuriteType.basal_dendrite)).max() for m in morphs],
    "Soma surface aerea": [features.morphology.soma_surface_area(m) for m in morphs],
    "Soma volumes": [features.morphology.soma_volume(m) for m in morphs],
    # "Neuron volumes": [sum(features.morphology.total_volume_per_neurite(m)) for m in morphs],
    "Neuron volume densities": [features.morphology.volume_density(m) for m in morphs],
    "Number of neurites": [features.morphology.number_of_neurites(m, neurite_type=NeuriteType.basal_dendrite) for m in morphs],
    "Total height": [features.morphology.total_height(m) for m in morphs]
}

# plot_side = int(ceil(sqrt(len(characteristics))))
fig, axs = plt.subplots(3, 4, figsize=(20, 12))
axs_flat = axs.ravel()

colors = plt.colormaps["Pastel1"](np.linspace(0, 0.99/3, 3))

for (k, vs), ax in zip(characteristics.items(), axs_flat):
    ax.bar(correct_m_names, vs, color=colors)
    ax.set_title(k)
    
plt.tight_layout()
```

```python
feature = "section_lengths"
f, ax = plt.subplots(1, len(morphs), figsize=(15, 5), sharex=True, sharey=True)
for i, morph in enumerate(morphs):
    ax[i].set_title(morph.name)
    # ax[i].set_xlim(1000)
    values = nm.get(feature, morph, neurite_type=nm.NeuriteType.axon)
    ax[i].hist(values,label=morph.name)
f.suptitle("Axonal section lenght")
f.supxlabel('Lenghts (um)')
f.supylabel("Density")
```
